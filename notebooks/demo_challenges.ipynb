# Résolution des challenges

## Utilisation

Soit **n** le numéro de challenge à résoudre. 

Si le challenge **n** est résolu :
1) Utiliser `solve_challenge(n)` pour résoudre le challenge avec ses paramètres par défaut
2) Utiliser `solve_challenge(n,**kwargs)` avec **kwargs** le dictionnaire avec les noms et valeurs des arguments
3) Utiliser `solve_challenge(n,*args)` avec **args** les arguments dans l'ordre du challenge

La fonction `get_kwargs(n)` permet d'obtenir un dictionnaire avec les arguments du challenge initialisés à leur valeur par défaut.\
La fonction `solve_challenge(n)` imprime toujours le readme, puis la solution et le temps d'exécution.
